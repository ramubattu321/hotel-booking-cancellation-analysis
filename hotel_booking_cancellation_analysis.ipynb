# Hotel Booking Data Wrangling & Platform Analysis

## Objective
This project focuses on cleaning semi-structured hotel booking data and identifying booking patterns across different platforms.

The goal is to transform messy raw data into a structured dataset suitable for platform-level analysis and reporting.

```python
df = pd.read_csv('hotel-booking-data.txt', delimiter='\t')
```

## Data Wrangling Logic
Some rows do not represent booking transactions. Instead, they contain booking platform markers such as Hotels, Booking, Expedia, or Cleartrip in the `Date` column.

These marker rows must be used to assign the correct booking source to the actual booking records.

```python
mask = df['Room number'].isna()
df['Booking Source'] = np.where(mask, df['Date'], np.nan)
```

```python
df[['Date', 'Room number', 'Booking Source']].head(10)
```

## Backfill Booking Source
Backfilling is used to assign each booking record to the nearest platform marker below it, allowing platform attribution for every valid reservation.

```python
df['Booking Source'].fillna(method='bfill', inplace=True)
```

```python
df[['Date', 'Room number', 'Booking Source']].head(10)
```

## Clean Final Dataset
Remove marker rows and incomplete records, keeping only valid booking transactions.

```python
df_clean = df.dropna(subset=['Company', 'Person Name', 'Room number']).copy()
```

```python
df_clean['Room number'] = df_clean['Room number'].astype(int)
```

```python
df_clean.head()
```

```python
df_clean.info()
```

## Booking Platform Distribution

```python
platform_counts = df_clean['Booking Source'].value_counts()
platform_counts
```

## Visualization

```python
plt.figure(figsize=(8, 5))
platform_counts.plot(
    kind='bar',
    title='Bookings by Platform',
    ylabel='Number of Bookings',
    xlabel='Booking Source',
    rot=45
)
plt.tight_layout()
plt.show()
```

## Key Insights
- Expedia generated the highest number of bookings in the dataset.
- Booking activity was distributed across multiple channels including Hotels, Booking, Cleartrip, and Travel Agent sources.
- Data wrangling was necessary to assign booking source information correctly.
- Cleaned data enabled accurate platform-level analysis from messy raw records.
